```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Bengaluru_House_Data.csv")
df.sample(5)
```

```python
df1 = df.drop(["availability" , "society" , "balcony", "area_type"] ,axis="columns")
```

```python
df1
```

```python
df1.isnull().sum()
df1.shape
```

```python
df2 = df1.dropna()
```

```python
df2.shape
```

```python
df2['size'].unique()
```

```python
df2['bhk'] = df2['size'].apply( lambda x : int(x.split(' ')[0]))
```

```python
df2
```

```python
df3 = df2.drop("size",axis = "columns")
```

```python
df3.sample(5)
```

```python
df3['total_sqft'].unique()
```

```python
def is_float(x):
    try:
        float(x)   
    except:
        return False
    return True 
```

```python
df3[~df3['total_sqft'].apply(is_float)]
```

```python
def convert_sqft_to_num(x):
    tokens = str(x).split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None   
```

```python
df4 = df3.copy()
df4.total_sqft = df4.total_sqft.apply(convert_sqft_to_num)
df4 = df4[df4.total_sqft.notnull()]
```

```python
df4
```

```python
df5 = df4.copy()
```

```python
df5['price_per_sqft'] = df5['price'] * 100000 / df5['total_sqft']
```

```python
df5
```

```python
df5["location"].unique().size
```

```python
df5.location = df5.location.apply( lambda x : x.strip())
```

```python
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
location_stats_less_than_10
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
df5
```

```python
df5['sqft_per_bedroom'] = df5['total_sqft'] / df5['bhk']
df5
```

```python
df5['sqft_per_bedroom'].std()
```

```python

df6 = df5[df5.sqft_per_bedroom > 300]
df6
```

```python
df6.price_per_sqft.describe()
```

```python
stddev = df6.price_per_sqft.std()
meandata =  df6.price_per_sqft.mean()
stddev
meandata
```

```python
min_threshhold = meandata - stddev
max_threshhold = meandata + stddev 
df7 = df6[ (df6.price_per_sqft > min_threshhold) & (df6.price_per_sqft < max_threshhold) ]
df7
```

```python
df7 = df7[ df7.bath <= df7.bhk + 2 ]
df7
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df8 = remove_bhk_outliers(df7)
# df8 = df7.copy()
df8.shape


df8
```

```python
df9 = df8.drop(['sqft_per_bedroom','price_per_sqft'],axis='columns')
df8 = df9
dummies = pd.get_dummies(df8.location)
dummies.head(3)
```

```python
df9 = pd.concat([df8,dummies.drop('other',axis='columns')],axis='columns')
df9 = df9.drop('location',axis='columns')
df9.head()
```

```python
X = df9.drop('price' ,axis='columns')
X
```

```python
y = df9['price']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)

from sklearn.linear_model import Lasso
L_clf = Lasso()
L_clf.fit(X_train, y_train)
L_clf.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['squared_error'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=4, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X_train,y_train)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return L_clf.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1000, 2, 2)
```
